## 2. Implementation of Core Methods

### 2.1 PCMCI Implementation for Causal Inference

```python
# Import Tigramite for PCMCI
from tigramite import data_processing as pp
from tigramite.pcmci import PCMCI
# from tigramite.independence_tests import ParCorr
from tigramite.independence_tests.parcorr import ParCorr

def estimate_effective_connectivity(data, tau_max=3, pc_alpha=0.05):
    """
    Estimate effective brain connectivity using PCMCI method.
    
    Parameters:
    -----------
    data : np.array
        Multivariate time series data (T x N) where T is time points and N is number of variables
    tau_max : int
        Maximum time lag to consider
    pc_alpha : float
        Significance level for PC algorithm
    
    Returns:
    --------
    W : np.array
        Weighted directed connectivity matrix
    """
    # Prepare data for Tigramite
    dataframe = pp.DataFrame(data)
    
    # Initialize PCMCI with partial correlation test
    parcorr = ParCorr(significance='analytic')
    pcmci = PCMCI(dataframe=dataframe, cond_ind_test=parcorr, verbosity=0)
    
    # Run PCMCI algorithm
    results = pcmci.run_pcmci(tau_max=tau_max, pc_alpha=pc_alpha)
    
    # Extract p-values and create weighted connectivity matrix
    p_matrix = results['p_matrix']
    n_vars = data.shape[1]
    W = np.zeros((n_vars, n_vars))
    
    # Average p-values across lags (Equation 2 in paper)
    for i in range(n_vars):
        for j in range(n_vars):
            if i != j:
                # Average p-values from j to i across all lags
                p_avg = np.mean(p_matrix[i, j, 1:tau_max+1])  
                W[i, j] = 1 - p_avg  # Convert p-value to weight (Equation 1)
    
    return W
```

### 2.2 Network Decomposition

```python
def decompose_network(W):
    """
    Decompose connectivity network into symmetric and anti-symmetric components.
    
    Parameters:
    -----------
    W : np.array
        Weighted connectivity matrix
    
    Returns:
    --------
    Ws : np.array
        Symmetric component (mutual influence)
    Wa : np.array
        Anti-symmetric component (net influence)
    """
    Ws = 0.5 * (W + W.T)  # Symmetric component
    Wa = 0.5 * (W - W.T)  # Anti-symmetric component
    
    return Ws, Wa
```

### 2.3 Hodge Decomposition Implementation

```python
class HodgeDecomposition:
    """
    Hodge Decomposition for ranking nodes in a directed network.
    Based on Jiang et al. (2011) - Statistical ranking and combinatorial Hodge theory
    """
    
    def __init__(self, n_nodes):
        self.n_nodes = n_nodes
        self.edges = [(i, j) for i in range(n_nodes) for j in range(i+1, n_nodes)]
        self.n_edges = len(self.edges)
        
    def create_gradient_matrix(self):
        """
        Create the gradient operator matrix that maps node potentials to edge flows.
        """
        rows, cols, data = [], [], []
        
        for edge_idx, (i, j) in enumerate(self.edges):
            # For edge (i,j): grad(s) = s_j - s_i
            rows.append(edge_idx)
            cols.append(i)
            data.append(-1)
            
            rows.append(edge_idx)
            cols.append(j)
            data.append(1)
        
        D = csr_matrix((data, (rows, cols)), shape=(self.n_edges, self.n_nodes))
        return D
    
    def vectorize_antisymmetric_matrix(self, Wa):
        """
        Convert anti-symmetric matrix to edge flow vector.
        """
        edge_flows = np.zeros(self.n_edges)
        
        for edge_idx, (i, j) in enumerate(self.edges):
            edge_flows[edge_idx] = Wa[i, j]
        
        return edge_flows
    
    def compute_ranking(self, Wa):
        """
        Compute node ranking scores using least squares (Equation 7 in paper).
        
        Parameters:
        -----------
        Wa : np.array
            Anti-symmetric component of connectivity matrix
        
        Returns:
        --------
        s : np.array
            Ranking scores for each node
        """
        # Create gradient matrix
        D = self.create_gradient_matrix()
        
        # Vectorize anti-symmetric matrix
        edge_flows = self.vectorize_antisymmetric_matrix(Wa)
        
        # Solve least squares: min ||Wa + grad(s)||^2
        # This is equivalent to solving: D^T * D * s = -D^T * edge_flows
        result = lsqr(D, -edge_flows)
        s = result[0]
        
        # Center the scores (remove mean)
        s = s - np.mean(s)
        
        return s
```




## Hodge Decomposition â€” Implementation (Math)

**Notation**
Let the graph have $N$ nodes and let $\mathcal{E}=\{e_1,\dots,e_M\}$ denote the set of undirected edges indexed by ordered node pairs $(i,j)$ with $i<j$. Let $W_a \in \mathbb{R}^{N \times N}$ be the anti-symmetric net-flow matrix so that $W_a^\top = -W_a$.

**Gradient (design) matrix**
Define the gradient matrix $D \in \mathbb{R}^{M \times N}$ with a row for each edge $e=(i,j) \in \mathcal{E}$:

$$
D_{e,p} =
\begin{cases}
-1, & \text{if } p=i, \\[4pt]
+1, & \text{if } p=j, \\[4pt]
0,  & \text{otherwise.}
\end{cases}
$$

For any node potential vector $s \in \mathbb{R}^N$, the induced edge flow (gradient) is:

$$
\mathrm{grad}(s) = D s, \qquad (\mathrm{grad}(s))_e = s_j - s_i \quad \text{for } e=(i,j).
$$

**Edge-flow vectorization**
Vectorize the anti-symmetric matrix into an edge-flow vector $f \in \mathbb{R}^M$ by:

$$
f_e = W_a[i,j] \quad \text{for } e=(i,j) \in \mathcal{E}.
$$

**Least-squares objective and normal equations**
Obtain the gradient component $\hat{s}$ by minimizing the edge residual:

$$
\hat{s} = \arg\min_{s \in \mathbb{R}^N} \; \| f + D s \|_2^2 .
$$

The normal equations are:

$$
D^\top D \, s = -D^\top f.
$$

Define $L \triangleq D^\top D$ (the graph Laplacian); $L$ is singular because constants lie in its kernel, so the solution is unique up to an additive constant.

**Regularized solve and centering**
For numerical stability one can solve the Tikhonov-regularized system:

$$
( D^\top D + \lambda I ) s = -D^\top f, \qquad \lambda \ge 0,
$$

and then fix the additive degree of freedom by centering:

$$
\hat{s} \leftarrow \hat{s} - \frac{1}{N}\mathbf{1}^\top \hat{s} .
$$

**Residual and interpretation**
The residual vector $r = f + D\hat{s}$ contains the non-gradient part of the flow (curl + harmonic). Its norm quantifies how far the data are from an exact gradient flow.



## 3. Simulation Experiments

### 3.1 Generate Simulation Data

```python
def create_mixing_matrix(scenario):
    """
    Create mixing matrix for different simulation scenarios.
    
    Scenarios:
    1: Linear hierarchy A->B->C->D->E
    2: Cyclic structure with all nodes connected
    3: Mixed structure with partial cycles
    4-6: Reversed versions of scenarios 1-3
    """
    Phi = np.zeros((5, 5))
    
    if scenario == 1:
        # Linear hierarchy: A->B->C->D->E
#         Phi[1, 0] = 0.4
#         Phi[2, 1] = 0.4
#         Phi[3, 2] = 0.4
#         Phi[4, 3] = 0.4
        Phi[1, 0] = 0.4  
        Phi[2, 0] = 0.5  
        Phi[2,1] = 0.4  
        Phi[3,1] = 0.5 
        Phi[3,2]= 0.4
        Phi[4,2]= 0.5
        Phi[4,3]=0.4
        
    elif scenario == 2:
        # Cyclic structure
        for i in range(5):
            for j in range(5):
                if i != j:
                    Phi[i, j] = 0.2
                    
    elif scenario == 3:
        # Mixed structure
        Phi[1, 0] = 0.4
        Phi[2, 1] = 0.4
        Phi[3, 2] = 0.3
        Phi[4, 3] = 0.3
        Phi[2, 4] = 0.2  # Create partial cycle
        
    elif scenario == 4:
        # Reversed linear hierarchy: E->D->C->B->A
        Phi=create_mixing_matrix(1).T
        
    elif scenario == 5:
        # Reversed cyclic (same as scenario 2)
        for i in range(5):
            for j in range(5):
                if i != j:
                    Phi[i, j] = 0.2
                    
    elif scenario == 6:
        # Reversed mixed structure
        Phi[3, 4] = 0.4
        Phi[2, 3] = 0.4
        Phi[1, 2] = 0.3
        Phi[0, 1] = 0.3
        Phi[4, 2] = 0.2  # Create partial cycle
    
    return Phi

def generate_simulation_data(scenario, T=5000):
    """
    Generate multivariate time series data for a given scenario.
    Using the model: Y(t) = Phi(t) * Z(t) + E(t)
    """
    P = 5  # Number of dimensions
    Phi = create_mixing_matrix(scenario)
    
    # Initialize data
    Y = np.zeros((T, P))
    
    # Generate time series with VAR(1) structure
    for t in range(1, T):
        # VAR(1) dynamics
        Y[t, :] = Phi @ Y[t-1, :] + np.random.randn(P)
    
    return Y
```

### 3.2 Run Simulation Experiments

```python
def run_simulation_experiment(scenario, n_epochs=10, epoch_length=500):
    """
    Run simulation experiment for a given scenario.
    """
    all_rankings = []
    
    for epoch in range(n_epochs):
        # Generate data for this epoch
        data = generate_simulation_data(scenario, T=epoch_length)
        
        # Estimate effective connectivity
        W = estimate_effective_connectivity(data, tau_max=2)
        
        # Decompose network
        Ws, Wa = decompose_network(W)
        
        # Compute rankings using Hodge decomposition
        hodge = HodgeDecomposition(n_nodes=5)
        rankings = hodge.compute_ranking(Wa)
        
        all_rankings.append(rankings)
    
    return np.array(all_rankings)

# Run experiments for all scenarios
print("Running simulation experiments...")
results = {}

for scenario in range(1, 7):
    print(f"\nScenario {scenario}:")
    rankings = run_simulation_experiment(scenario)
    results[f'Scenario {scenario}'] = rankings
    
    # Print average rankings
    avg_rankings = np.mean(rankings, axis=0)
    node_labels = ['A', 'B', 'C', 'D', 'E']
    ranking_order = np.argsort(-avg_rankings)  # Sort in descending order
    
    print("Average node rankings (highest to lowest):")
    for idx in ranking_order:
        print(f"  {node_labels[idx]}: {avg_rankings[idx]:.3f}")
```

```python
"""
Topological Analysis of Seizure-Induced Changes in Brain Hierarchy
Complete replication pipeline with proper PCMCI and first differencing
"""

import numpy as np
import pandas as pd
import mne
import matplotlib.pyplot as plt
from pathlib import Path
import re
import warnings
from scipy import linalg
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import lsqr
from scipy.stats import mannwhitneyu
import seaborn as sns

# Tigramite imports for PCMCI
# from tigramite import data_processing as pp
# from tigramite.pcmci import PCMCI
# from tigramite.independence_tests import ParCorr

warnings.filterwarnings("ignore")


# ======================== CONFIGURATION ========================
edf_path = Path("../data/eeg1.edf")
annot_paths = [
    Path("../data/annotations_2017_B.csv"),
    Path("../data/annotations_2017_C.csv")
]
# Select channels to analyze (paper uses 19 channels)
desired_channels = ["F3", "F7", "Fz", "C3", "C4", "Cz", "P3", "P4", "Pz", 
                   "O1", "O2", "T3", "T4", "T5", "T6", "Fp1", "Fp2", "F4", "F8"]
output_dir = Path("../data/results")
output_dir.mkdir(exist_ok=True)

# Analysis parameters
N_SAMPLE = 20  # Start with 20, increase to 50 once verified
TAU_MAX = 3    # Maximum time lag
PC_ALPHA = 0.05  # Significance level
DOWNSAMPLE_FACTOR = 5  # Downsample for speed (EEG is often oversampled)


# ======================== HELPER FUNCTIONS ========================

def estimate_effective_connectivity(epoch_data, tau_max=3, pc_alpha=0.05, 
                                   use_first_diff=True, downsample=5):
    """
    Estimate effective connectivity using PCMCI with first differencing.
    
    Parameters:
    -----------
    epoch_data : ndarray, shape (n_channels, n_times)
        Single epoch EEG data
    tau_max : int
        Maximum time lag to consider
    pc_alpha : float
        Significance level for PC algorithm
    use_first_diff : bool
        Apply first differencing (CRITICAL for EEG!)
    downsample : int
        Downsample factor for speed
    
    Returns:
    --------
    W : ndarray, shape (n_channels, n_channels)
        Weighted directed connectivity matrix
    """
    n_channels, n_times = epoch_data.shape
    
    # Downsample for speed
    if downsample > 1 and n_times > 500:
        epoch_data = epoch_data[:, ::downsample]
        n_times = epoch_data.shape[1]
    
    # CRITICAL: Apply first differencing to reduce autocorrelation
    # This is explicitly mentioned in the paper
    if use_first_diff:
        data = np.diff(epoch_data, axis=1).T  # Shape: (n_times-1, n_channels)
    else:
        data = epoch_data.T  # Shape: (n_times, n_channels)
    
    # Standardize each channel
    data = (data - np.mean(data, axis=0)) / (np.std(data, axis=0) + 1e-10)
    
    try:
        # Prepare data for Tigramite
        dataframe = pp.DataFrame(data)
        
        # Initialize PCMCI with partial correlation test
        parcorr = ParCorr(significance='analytic')
        pcmci = PCMCI(dataframe=dataframe, cond_ind_test=parcorr, verbosity=0)
        
        # Run PCMCI algorithm
        results = pcmci.run_pcmci(tau_max=tau_max, pc_alpha=pc_alpha)
        
        # Extract p-values and create weighted connectivity matrix
        p_matrix = results['p_matrix']
        W = np.zeros((n_channels, n_channels))
        
        # Average p-values across lags (Equation 2 in paper)
        for i in range(n_channels):
            for j in range(n_channels):
                if i != j:
                    # Average p-values from j to i across all lags (excluding lag 0)
                    p_vals = p_matrix[i, j, 1:tau_max+1]
                    # Filter out any NaN or inf values
                    p_vals = p_vals[np.isfinite(p_vals)]
                    if len(p_vals) > 0:
                        p_avg = np.mean(p_vals)
                        W[i, j] = 1 - p_avg  # Convert p-value to weight (Equation 1)
        
    except Exception as e:
        print(f"    [Warning] PCMCI failed: {e}, using fallback method")
        # Fallback: simple time-lagged correlation
        W = fallback_connectivity(data, tau_max)
    
    return W


def fallback_connectivity(data, tau_max=3):
    """
    Fallback connectivity estimation using time-lagged correlation.
    Used if PCMCI fails or for quick testing.
    """
    n_channels = data.shape[1]
    W = np.zeros((n_channels, n_channels))
    
    for i in range(n_channels):
        for j in range(n_channels):
            if i != j:
                correlations = []
                for tau in range(1, tau_max + 1):
                    if tau < data.shape[0]:
                        corr = np.corrcoef(data[tau:, i], data[:-tau, j])[0, 1]
                        if not np.isnan(corr):
                            correlations.append(abs(corr))
                W[i, j] = np.mean(correlations) if correlations else 0
    
    return W


def decompose_network(W):
    """
    Decompose network into symmetric and antisymmetric components.
    
    Ws: reciprocal connections (non-hierarchical)
    Wa: hierarchical flow (used for ranking)
    """
    Ws = 0.5 * (W + W.T)
    Wa = 0.5 * (W - W.T)
    return Ws, Wa


class HodgeDecomposition:
    """
    Hodge Decomposition for ranking nodes in a directed network.
    Based on Jiang et al. (2011) - Statistical ranking and combinatorial Hodge theory
    """
    
    def __init__(self, n_nodes):
        self.n_nodes = n_nodes
        self.edges = [(i, j) for i in range(n_nodes) for j in range(i+1, n_nodes)]
        self.n_edges = len(self.edges)
        
    def create_gradient_matrix(self):
        """
        Create the gradient operator matrix that maps node potentials to edge flows.
        For edge (i,j): grad(s) = s_j - s_i
        """
        rows, cols, data = [], [], []
        
        for edge_idx, (i, j) in enumerate(self.edges):
            rows.append(edge_idx)
            cols.append(i)
            data.append(-1)
            
            rows.append(edge_idx)
            cols.append(j)
            data.append(1)
        
        D = csr_matrix((data, (rows, cols)), shape=(self.n_edges, self.n_nodes))
        return D
    
    def vectorize_antisymmetric_matrix(self, Wa):
        """
        Convert anti-symmetric matrix to edge flow vector.
        """
        edge_flows = np.zeros(self.n_edges)
        
        for edge_idx, (i, j) in enumerate(self.edges):
            edge_flows[edge_idx] = Wa[i, j]
        
        return edge_flows
    
    def compute_ranking(self, Wa):
        """
        Compute node ranking scores using least squares (Equation 7 in paper).
        
        Solves: min ||Wa + grad(s)||^2
        Equivalent to: D^T * D * s = -D^T * edge_flows
        
        Returns:
        --------
        s : ndarray
            Ranking scores for each node (centered)
        """
        # Create gradient matrix
        D = self.create_gradient_matrix()
        
        # Vectorize anti-symmetric matrix
        edge_flows = self.vectorize_antisymmetric_matrix(Wa)
        
        # Solve least squares
        result = lsqr(D, -edge_flows)
        s = result[0]
        
        # Center the scores (remove mean)
        s = s - np.mean(s)
        
        return s


def analyze_epoch(epoch_data, channel_names, tau_max=3, pc_alpha=0.05):
    """
    Analyze a single epoch: connectivity -> decomposition -> ranking.
    
    Parameters:
    -----------
    epoch_data : ndarray, shape (n_channels, n_times)
        Single epoch EEG data
    channel_names : list
        Names of channels
    tau_max : int
        Maximum time lag for connectivity
    pc_alpha : float
        Significance level for PCMCI
    
    Returns:
    --------
    results : dict
        Contains W, Ws, Wa, and rankings
    """
    n_channels = epoch_data.shape[0]
    
    # Estimate effective connectivity with PCMCI + first differencing
    W = estimate_effective_connectivity(epoch_data, tau_max=tau_max, 
                                       pc_alpha=pc_alpha, 
                                       use_first_diff=True,
                                       downsample=DOWNSAMPLE_FACTOR)
    
    # Decompose network
    Ws, Wa = decompose_network(W)
    
    # Compute rankings using Hodge decomposition
    hodge = HodgeDecomposition(n_channels)
    rankings = hodge.compute_ranking(Wa)
    
    return {
        'W': W,
        'Ws': Ws,
        'Wa': Wa,
        'rankings': rankings,
        'channel_names': channel_names
    }


# ======================== MAIN PIPELINE ========================

def main():
    print("="*70)
    print("EEG SEIZURE ANALYSIS - TOPOLOGICAL PIPELINE (PCMCI + Hodge)")
    print("="*70)
    
    # 1) Load EDF
    print("\n[1/7] Loading EEG data...")
    raw = mne.io.read_raw_edf(str(edf_path), preload=True, verbose=False)
    sfreq = raw.info['sfreq']
    total_seconds = raw.n_times / sfreq
    print(f"  Sampling freq: {sfreq} Hz")
    print(f"  Total duration: {total_seconds:.1f} s")
    print(f"  Available channels: {len(raw.ch_names)}")
    
    # 2) Load annotations
    print("\n[2/7] Loading annotations...")
    annot_df = None
    for p in annot_paths:
        if p.exists():
            print(f"  Found: {p}")
            try:
                annot_df = pd.read_csv(p, header=0)
            except:
                annot_df = pd.read_csv(p, header=None)
            break
    
    if annot_df is None:
        raise FileNotFoundError("No annotation CSV found!")
    
    # Determine column for this EDF
    basename = edf_path.name
    m = re.search(r"(\d+)", basename)
    candidate_col = None
    if m:
        colnum = m.group(1)
        for col in annot_df.columns:
            if str(col).strip() == colnum or str(col).strip().lower() == f"eeg{colnum}":
                candidate_col = col
                break
    if candidate_col is None:
        candidate_col = annot_df.columns[0]
    
    labels = annot_df[candidate_col].astype(float).values
    n_epochs_annot = len(labels)
    epoch_length_s = total_seconds / n_epochs_annot
    print(f"  Annotations: {n_epochs_annot} epochs")
    print(f"  Epoch length: {epoch_length_s:.4f} s")
    
    # 3) Create fixed-length epochs
    print("\n[3/7] Creating epochs...")
    epochs = mne.make_fixed_length_epochs(raw, duration=epoch_length_s, 
                                         preload=True, verbose=False)
    data = epochs.get_data()  # shape: (n_epochs, n_channels, n_times)
    print(f"  Epochs created: {data.shape}")
    
    # 4) Select channels
    print("\n[4/7] Selecting channels...")
    picked_channels = []
    for target in desired_channels:
        found = next((ch for ch in raw.ch_names if target.lower() in ch.lower()), None)
        if found:
            picked_channels.append(found)
    
    print(f"  Selected {len(picked_channels)} channels:")
    for i, ch in enumerate(picked_channels[:5]):
        print(f"    {i}: {ch}")
    if len(picked_channels) > 5:
        print(f"    ... and {len(picked_channels)-5} more")
    
    if len(picked_channels) < 3:
        raise ValueError("Need at least 3 channels for analysis!")
    
    # Get channel indices
    chan_indices = [raw.ch_names.index(ch) for ch in picked_channels]
    data_selected = data[:, chan_indices, :]
    
    # Align labels
    n_epochs = data_selected.shape[0]
    L = min(n_epochs, len(labels))
    labels = labels[:L]
    data_selected = data_selected[:L]
    
    # 5) Separate seizure and non-seizure epochs
    print("\n[5/7] Separating seizure and non-seizure epochs...")
    seizure_indices = np.where(labels == 1)[0]
    non_seizure_indices = np.where(labels == 0)[0]
    
    print(f"  Seizure epochs: {len(seizure_indices)}")
    print(f"  Non-seizure epochs: {len(non_seizure_indices)}")
    
    # Sample epochs for analysis
    n_sample = min(N_SAMPLE, len(seizure_indices), len(non_seizure_indices))
    
    np.random.seed(42)
    seizure_sample = np.random.choice(seizure_indices, size=n_sample, replace=False)
    non_seizure_sample = np.random.choice(non_seizure_indices, size=n_sample, replace=False)
    
    print(f"  Analyzing {n_sample} epochs of each type")
    print(f"  [Note] With PCMCI, estimated runtime: ~{n_sample * 2 * 15 / 60:.1f} minutes")
    
    # 6) Run topological analysis on each epoch
    print("\n[6/7] Running topological analysis...")
    print(f"  Using: PCMCI (tau_max={TAU_MAX}) + First Differencing + Hodge Decomposition")
    
    import time
    start_time = time.time()
    
    seizure_rankings = []
    non_seizure_rankings = []
    
    print("  Processing seizure epochs...")
    for i, idx in enumerate(seizure_sample):
        epoch_data = data_selected[idx]  # shape: (n_channels, n_times)
        result = analyze_epoch(epoch_data, picked_channels, 
                              tau_max=TAU_MAX, pc_alpha=PC_ALPHA)
        seizure_rankings.append(result['rankings'])
        
        if (i + 1) % 5 == 0 or i == 0:
            elapsed = time.time() - start_time
            eta = elapsed / (i + 1) * (2 * n_sample - i - 1)
            print(f"    {i+1}/{n_sample} done | Elapsed: {elapsed/60:.1f}m | ETA: {eta/60:.1f}m")
    
    print("  Processing non-seizure epochs...")
    for i, idx in enumerate(non_seizure_sample):
        epoch_data = data_selected[idx]
        result = analyze_epoch(epoch_data, picked_channels,
                              tau_max=TAU_MAX, pc_alpha=PC_ALPHA)
        non_seizure_rankings.append(result['rankings'])
        
        if (i + 1) % 5 == 0:
            elapsed = time.time() - start_time
            eta = elapsed / (n_sample + i + 1) * (n_sample - i - 1)
            print(f"    {i+1}/{n_sample} done | Elapsed: {elapsed/60:.1f}m | ETA: {eta/60:.1f}m")
    
    total_time = time.time() - start_time
    print(f"  âœ“ Analysis complete! Total time: {total_time/60:.1f} minutes")
    
    seizure_rankings = np.array(seizure_rankings)  # shape: (n_sample, n_channels)
    non_seizure_rankings = np.array(non_seizure_rankings)
    
    # 7) Statistical comparison and visualization
    print("\n[7/7] Analyzing results and creating visualizations...")
    
    # Compute statistics for each channel
    channel_stats = []
    for ch_idx, ch_name in enumerate(picked_channels):
        seiz_scores = seizure_rankings[:, ch_idx]
        non_seiz_scores = non_seizure_rankings[:, ch_idx]
        
        # Mann-Whitney U test
        stat, pval = mannwhitneyu(seiz_scores, non_seiz_scores, alternative='two-sided')
        
        channel_stats.append({
            'channel': ch_name.replace('EEG ', '').replace('-REF', '').strip(),
            'seizure_mean': np.mean(seiz_scores),
            'seizure_std': np.std(seiz_scores),
            'non_seizure_mean': np.mean(non_seiz_scores),
            'non_seizure_std': np.std(non_seiz_scores),
            'mean_diff': np.mean(seiz_scores) - np.mean(non_seiz_scores),
            'p_value': pval
        })
    
    stats_df = pd.DataFrame(channel_stats)
    stats_df = stats_df.sort_values('p_value')
    
    print("\n" + "="*70)
    print("RESULTS: Channels with significant ranking score differences")
    print("="*70)
    print(stats_df.to_string(index=False))
    
    # Save results
    stats_df.to_csv(output_dir / "channel_statistics.csv", index=False)
    
    # Visualization 1: Box plots for most significant channels (like Figure 8)
    n_top = min(6, len(picked_channels))
    top_channels = stats_df.head(n_top)
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    axes = axes.flatten()
    
    for i, (_, row) in enumerate(top_channels.iterrows()):
        if i >= n_top:
            break
        
        ch_name = row['channel']
        ch_idx = [j for j, ch in enumerate(picked_channels) 
                  if ch_name in ch.replace('EEG ', '').replace('-REF', '').strip()][0]
        
        seiz_data = seizure_rankings[:, ch_idx]
        non_seiz_data = non_seizure_rankings[:, ch_idx]
        
        data_plot = [seiz_data, non_seiz_data]
        labels_plot = ['Seizure', 'No Seizure']
        
        bp = axes[i].boxplot(data_plot, labels=labels_plot, patch_artist=True)
        for patch in bp['boxes']:
            patch.set_facecolor('lightblue')
        
        axes[i].set_title(f'Channel {ch_name}')
        axes[i].set_ylabel('Ranking Score')
        axes[i].grid(True, alpha=0.3)
        
        # Add p-value and mean difference
        axes[i].text(0.5, 0.95, f"Î”={row['mean_diff']:.4f}\np={row['p_value']:.4f}",
                    transform=axes[i].transAxes, ha='center', va='top',
                    bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
                    fontsize=8)
    
    plt.tight_layout()
    plt.savefig(output_dir / "ranking_comparison_boxplots.png", dpi=200, bbox_inches='tight')
    print(f"\nâœ“ Saved: {output_dir / 'ranking_comparison_boxplots.png'}")
    
    # Visualization 2: Heatmap of ranking scores
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    vmax = max(abs(seizure_rankings.min()), abs(seizure_rankings.max()),
               abs(non_seizure_rankings.min()), abs(non_seizure_rankings.max()))
    
    im1 = ax1.imshow(seizure_rankings.T, aspect='auto', cmap='RdBu_r', 
                     interpolation='nearest', vmin=-vmax, vmax=vmax)
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Channel')
    ax1.set_title('Seizure Epochs - Ranking Scores')
    ax1.set_yticks(range(len(picked_channels)))
    ax1.set_yticklabels([ch.replace('EEG ', '').replace('-REF', '').strip() 
                         for ch in picked_channels], fontsize=7)
    plt.colorbar(im1, ax=ax1)
    
    im2 = ax2.imshow(non_seizure_rankings.T, aspect='auto', cmap='RdBu_r',
                     interpolation='nearest', vmin=-vmax, vmax=vmax)
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Channel')
    ax2.set_title('Non-Seizure Epochs - Ranking Scores')
    ax2.set_yticks(range(len(picked_channels)))
    ax2.set_yticklabels([ch.replace('EEG ', '').replace('-REF', '').strip() 
                         for ch in picked_channels], fontsize=7)
    plt.colorbar(im2, ax=ax2)
    
    plt.tight_layout()
    plt.savefig(output_dir / "ranking_heatmaps.png", dpi=200, bbox_inches='tight')
    print(f"âœ“ Saved: {output_dir / 'ranking_heatmaps.png'}")
    
    # Visualization 3: Mean difference plot
    fig, ax = plt.subplots(figsize=(12, 6))
    
    mean_diffs = [row['mean_diff'] for _, row in stats_df.iterrows()]
    p_values = [row['p_value'] for _, row in stats_df.iterrows()]
    channel_labels = [row['channel'] for _, row in stats_df.iterrows()]
    
    colors = ['red' if p < 0.05 else 'gray' for p in p_values]
    
    bars = ax.barh(range(len(channel_labels)), mean_diffs, color=colors, alpha=0.7)
    ax.set_yticks(range(len(channel_labels)))
    ax.set_yticklabels(channel_labels, fontsize=8)
    ax.set_xlabel('Mean Ranking Score Difference (Seizure - Non-Seizure)')
    ax.set_title('Channel-wise Ranking Score Changes During Seizures\n(Red = p < 0.05)')
    ax.axvline(x=0, color='black', linestyle='--', linewidth=1)
    ax.grid(True, alpha=0.3, axis='x')
    
    plt.tight_layout()
    plt.savefig(output_dir / "mean_difference_plot.png", dpi=200, bbox_inches='tight')
    print(f"âœ“ Saved: {output_dir / 'mean_difference_plot.png'}")
    
    print("\n" + "="*70)
    print("ANALYSIS COMPLETE!")
    print("="*70)
    print(f"Results saved to: {output_dir}")
    print(f"\nKey findings:")
    print(f"  - {len(stats_df[stats_df['p_value'] < 0.05])} channels show significant changes (p < 0.05)")
    print(f"  - Most affected channels:")
    for _, row in stats_df.head(3).iterrows():
        print(f"    â€¢ {row['channel']}: Î”={row['mean_diff']:.4f}, p={row['p_value']:.6f}")
    
    print(f"\nMethodology:")
    print(f"  âœ“ First differencing applied (reduces autocorrelation)")
    print(f"  âœ“ PCMCI for causal connectivity (tau_max={TAU_MAX})")
    print(f"  âœ“ Hodge decomposition for hierarchical ranking")
    print(f"  âœ“ Mann-Whitney U test for statistical comparison")
    
    return {
        'seizure_rankings': seizure_rankings,
        'non_seizure_rankings': non_seizure_rankings,
        'statistics': stats_df,
        'channel_names': picked_channels
    }


if __name__ == "__main__":
    results = main()
    plt.show()
```
