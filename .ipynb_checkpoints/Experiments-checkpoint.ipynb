```python
import numpy as np
import matplotlib.pyplot as plt
from tigramite import data_processing as pp
from tigramite.pcmci import PCMCI
from tigramite.independence_tests.parcorr import ParCorr
#from tigramite.independence_tests import ParCorr
from scipy.linalg import lstsq
from scipy.stats import multivariate_normal  # For Z(t), E(t)
np.random.seed(0) 

# Node labels
nodes = ['A', 'B', 'C', 'D', 'E']
n_nodes = 5
T_total = 30000  # Total obs
num_epochs = 60
epoch_len = T_total // num_epochs  # 600

# Parameters
tau_max = 1  # Small for short epochs (adjust if needed)
pc_alpha = 0.05
p_threshold = 0.05
aggregate_p = 'min'  # As in code
```

```python
def get_phi(scenario):
    phi = np.zeros((n_nodes, n_nodes))  # φ(t) fixed per scenario
    
    if scenario == 1:  # Linear: A→B→C→D→E
        phi[0, 1] = 0.4  
        phi[0, 2] = 0.5  
        phi[1, 2] = 0.4  
        phi[1, 3] = 0.5 
        phi[2, 3]= 0.4
        phi[2,4]= 0.5
        phi[3,4]=0.4
    elif scenario == 2:  # Cyclic: Full cycle A→B→C→D→E→A
        phi[1, 0] = 0.4
        phi[2, 1] = 0.5
        phi[3, 2] = 0.4
        phi[4, 3] = 0.5
        phi[0, 4] = 0.4  # Close cycle
    elif scenario == 3:  # Mixed: Linear + some cross
        phi[1, 0] = 0.4
        phi[2, 1] = 0.5
        phi[3, 2] = 0.4
        phi[4, 3] = 0.5
        phi[2, 0] = 0.3  # Extra A→C
        phi[4, 1] = 0.3  # Extra B→E
    elif scenario == 4:  # Reverse 1: E→D→C→B→A
        phi[3, 4] = 0.4  # D from E
        phi[2, 3] = 0.5  # C from D
        phi[1, 2] = 0.4  # B from C
        phi[0, 1] = 0.5  # A from B
    elif scenario == 5:  # Reverse 2: Reverse cycle
        phi[3, 4] = 0.4
        phi[2, 3] = 0.5
        phi[1, 2] = 0.4
        phi[0, 1] = 0.5
        phi[4, 0] = 0.4  # Close reverse cycle
    elif scenario == 6:  # Reverse 3: Reverse mixed
        phi[3, 4] = 0.4
        phi[2, 3] = 0.5
        phi[1, 2] = 0.4
        phi[0, 1] = 0.5
        phi[2, 4] = 0.3  # Extra E→C
        phi[4, 2] = 0.3  # Extra C→E? Wait, reverse B→E to E→B? Adjust per fig.
    else:
        raise ValueError(f"Unknown scenario {scenario}")
    
    return phi
```

```python
def generate_data(scenario, T=5000):
    phi = get_phi(scenario)
    Y = np.zeros((n_nodes, T))
    cov = np.eye(n_nodes)  # Identity for Z, E
    
    for t in range(T):
        Z = multivariate_normal.rvs(mean=np.zeros(n_nodes), cov=cov)
        E = multivariate_normal.rvs(mean=np.zeros(n_nodes), cov=cov)
        Y[:, t] = phi @ Z + E  # Y(t) = φ Z(t) + E(t)
    
    return Y

def generate_var1(T=5000, noise_std=1.0):
    Phi = np.zeros((5,5))
    # A->B, B->C, C->D, D->E, plus A->C as in paper
    Phi[1,0] = 0.4  # A->B
    Phi[2,1] = 0.4  # B->C
    Phi[3,2] = 0.4  # C->D
    Phi[4,3] = 0.4  # D->E
    Phi[2,0] = 0.5  # A->C (secondary)
#     Phi= (get_phi(1)).T
    Y = np.zeros((5, T))
    eps = np.random.normal(0, noise_std, size=(5, T))
    for t in range(1, T):
        Y[:, t] = Phi @ Y[:, t-1] + eps[:, t]
    return Y
```

```python
def cbtr_on_epochs(Y, ch_names=None, epoch_len=500, **pcmci_kwargs):
    """
    Run CBTR on epochs of Y, return average ranking scores.
    - Y: array-like shape (n_channels, T)
    - ch_names: list of channel names (or None)
    - epoch_len: epoch length in timepoints
    pcmci_kwargs may include: tau_max (default 1), pc_alpha (default 0.05),
    p_threshold (default 0.05).
    """
    n_channels, T = Y.shape
    num_epochs = T // epoch_len
    if num_epochs == 0:
        raise ValueError("T must be at least epoch_len")
    # reshape into (n_channels, num_epochs, epoch_len)
    epochs = Y[:, :num_epochs * epoch_len].reshape(n_channels, num_epochs, epoch_len)
    
    tau_max = pcmci_kwargs.get('tau_max', 1)
    pc_alpha = pcmci_kwargs.get('pc_alpha', 0.05)
    p_threshold = pcmci_kwargs.get('p_threshold', 0.05)
    
    s_hats = []
    for i in range(num_epochs):
        epoch_ts = epochs[:, i, :].T          # shape (epoch_len, n_channels)
        dataframe = pp.DataFrame(epoch_ts, var_names=ch_names)
        pcmci = PCMCI(dataframe=dataframe, cond_ind_test=ParCorr())
        results = pcmci.run_pcmci(tau_max=tau_max, pc_alpha=pc_alpha)
        p_matrix = results['p_matrix']       # shape (n_vars, n_vars, tau_max+1)
        
        # FIX 1: mean aggregation across lags (Eq. 2 in paper)
        p_agg = np.mean(p_matrix, axis=2) # p_agg[i,j] = p-value for i -> j averaged over lags
        print(f"epoch {i} p_agg (rounded):\n", np.round(p_agg, 3))
        
        # FIX 2: orientation per paper: W[p,q] = 1 - p_{q -> p}
        # so compute W_full = (1 - p_agg).T and only keep significant links (p < threshold)
        W_full = (1.0 - p_agg).T
        significance_mask = (p_agg.T < p_threshold) & (p_agg.T > 0)
        W = np.zeros_like(W_full)
        W[significance_mask] = W_full[significance_mask]
        
        # Symmetric and anti-symmetric parts
        W_s = (W + W.T) / 2.0
        W_a = (W - W.T) / 2.0
        print("W_a (rounded):\n", np.round(W_a,3))
        
        # FIX 3: Use only positive net flows (W_a > 0) and set incidence matrix with:
        #   +1 at source, -1 at target, and solve B s = f where f = W_a[target,source]
        edges = np.argwhere(W_a > 0.0)   # each row is (target p, source q)
        if len(edges) == 0:
            s_hats.append(np.zeros(n_channels))
            continue
        
        n_edges = len(edges)
        B = np.zeros((n_edges, n_channels))
        f = np.zeros(n_edges)
        for k, (p, q) in enumerate(edges):
            # W_a[p,q] > 0 indicates net flow from q -> p
            B[k, q] = +1.0   # source
            B[k, p] = -1.0   # target
            f[k] = W_a[p, q] # observed net flow magnitude (positive)
        
        # solve least-squares B s = f  (no extra negative sign)
        s_hat, *_ = np.linalg.lstsq(B, f, rcond=None)
        # center the score (paper centers / zero-mean)
        s_hat = s_hat - np.mean(s_hat)
        s_hats.append(s_hat)
    
    s_avg = np.mean(s_hats, axis=0)
    # re-center final average to zero mean (optional but matches earlier behavior)
    s_avg = s_avg - np.mean(s_avg)
    return s_avg
```

```python
scenarios = [1]#,2,3,4,5,6]
s_results = {}

for scen in scenarios:
    print(f"Running Scenario {scen}...")
    Y = generate_var1()
    s_avg = cbtr_on_epochs(Y, **{'tau_max':1, 'pc_alpha':0.05, 'p_threshold':0.05})
    s_results[scen] = s_avg
    print(f"Scenario {scen} average s: {s_avg}")

# # Plot aggregate rankings (Fig. 6)
# fig, ax = plt.subplots(figsize=(10, 6))
# for scen in scenarios:
#     ax.plot(range(1,7), s_results[scen], marker='o', label=f'Scenario {scen}')
# ax.set_xlabel('Epochs (Aggregated)')
# ax.set_ylabel('Ranking Scores')
# ax.set_title('Aggregate CBTR Rankings Across Scenarios')
# ax.legend()
# ax.grid(True)
# plt.show()
```

```python
# Plot aggregate rankings (Fig. 6)
fig, ax = plt.subplots(figsize=(10, 6))
for scen in scenarios:
    ax.plot(range(1,6), s_results[scen], marker='o', label=f'Scenario {scen}')
ax.set_xlabel('Epochs (Aggregated)')
ax.set_ylabel('Ranking Scores')
ax.set_title('Aggregate CBTR Rankings Across Scenarios')
ax.legend()
ax.grid(True)
plt.show()
```
